```python
import heapq #

def ucs(graph, start, goal):
    # Priority queue (cost, node, path)
    queue = [(0, start, [])]

    # Visited dictionary to store lowest cost
    visited = {}

    while queue:
        cost, node, path = heapq.heappop(queue)

        # If node already visited with lower cost, skip
        if node in visited and visited[node] <= cost:
            continue

        # Mark as visited
        visited[node] = cost

        # Update path
        path = path + [node]

        # Goal check
        if node == goal:
            return cost, path

        # Explore neighbors
        for neighbor, weight in graph[node]:
            heapq.heappush(queue, (cost + weight, neighbor, path))

    return float("inf"), []


# Graph based on your input
graph = {
    'S': [('A', 3), ('B', 2), ('C', 7)],
    'A': [('D', 3), ('E', 8), ('G', 15)],
    'B': [('G', 20)],
    'C': [('G', 6)],
    'D': [],
    'E': [],
    'G': []
}

# Run UCS
cost, path = ucs(graph, 'S', 'G')

# Output
print("Shortest Path:", " -> ".join(path))
print("Total Cost:", cost)
```

```python
import heapq   # Import heapq to use priority queue (min-heap)

# Function for Uniform Cost Search
def ucs(graph, start, goal):

    # Priority queue: stores tuples (cost, current_node, path_taken)
    queue = [(0, start, [])]   # Start with cost 0, starting node, empty path

    # Dictionary to store the minimum cost at which a node was visited
    visited = {}

    # Loop runs until queue becomes empty
    while queue:

        # Remove and return the element with smallest cost
        cost, node, path = heapq.heappop(queue)

        # If node already visited with a lower cost, skip it
        if node in visited and visited[node] <= cost:
            continue

        # Store the cost of visiting this node
        visited[node] = cost

        # Add current node to the path
        path = path + [node]

        # If goal node is reached, return result
        if node == goal:
            return cost, path

        # Traverse all neighbors of current node
        for neighbor, weight in graph[node]:

            # Add neighbor to queue with updated cost
            heapq.heappush(queue, (cost + weight, neighbor, path))

    # If goal is not found
    return float("inf"), []


# ---------------- GRAPH ----------------

# Graph represented as adjacency list
graph = {
    'S': [('A', 3), ('B', 2), ('C', 7)],   # From S to A, B, C
    'A': [('D', 3), ('E', 8), ('G', 15)],  # From A to D, E, G
    'B': [('G', 20)],                      # From B to G
    'C': [('G', 6)],                       # From C to G
    'D': [],                               # No outgoing edges
    'E': [],                               # No outgoing edges
    'G': []                                # Goal node
}

# ---------------- RUN UCS ----------------

# Call UCS function from start node S to goal node G
cost, path = ucs(graph, 'S', 'G')

# Print shortest path
print("Shortest Path:", " -> ".join(path))

# Print total cost of that path
print("Total Cost:", cost)
```

```python
import heapq

def ucs(graph, start, goal):
    # Priority queue (cost, node, path)
    queue = [(0, start, [])]

    # Visited dictionary to store lowest cost
    visited = {}

    while queue:
        cost, node, path = heapq.heappop(queue)

        # If node already visited with lower cost, skip
        if node in visited and visited[node] <= cost:
            continue

        # Mark as visited
        visited[node] = cost

        # Update path
        path = path + [node]

        # Goal check
        if node == goal:
            return cost, path

        # Explore neighbors
        for neighbor, weight in graph[node]:
            heapq.heappush(queue, (cost + weight, neighbor, path))

    return float("inf"), []


# Graph based on your input
graph = {
    'S': [('A', 3), ('B', 2), ('C', 7)],
    'A': [('D', 3), ('E', 8), ('G', 15)],
    'B': [('G', 20)],
    'C': [('G', 6)],
    'D': [],
    'E': [],
    'G': []
}

# Run UCS
cost, path = ucs(graph, 'S', 'G')

# Output
print("Shortest Path:", " -> ".join(path))
print("Total Cost:", cost)
```
